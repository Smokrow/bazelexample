# Terminal Cheat Sheet

## Basics 
### File and Directory Management

- List files and directories:
    ``` shell
     ls
    ```



- List all files and directories even if they are hidden:
    ``` shell
     ls -a
    ```
    
    
- Change directory:
    ``` shell
     cd [directory_name]
    ```


- Move to parent directory:
    ``` shell
     cd ..
    ```


- Create a new directory:
    ``` shell
     mkdir [directory_name]
    ```
    
    
- Remove a directory:
    ``` shell
     rmdir [directory_name]
    ```
    
    
- Create a new File:
    ``` shell
     touch [file_name]
    ```
    
    
- Remove a file:
    ``` shell
     rm [file_name]
    ```
    
    
- Copy a file:
    ``` shell
     cp [source][destination]
    ```
    
    
- Move or rename a file:
    ``` shell
     mv [source][destination]
    ```
 
 
 - View file content:
    ``` shell
     cat [file_name]
    ```


 - Clear the terminal screen :
    ``` shell
     clear
    ```
